### seaborn
https://seaborn.pydata.org/introduction.html

https://seaborn.pydata.org/generated/seaborn.lmplot.html?highlight=seaborn%20lmplot#seaborn.lmplot

```python
sns.lmplot(x='math_score', y='reading_score', data=students_performance) #график из библиотеки seaborn
```

добавим группирующую переменную:

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance) #
```

без регрессионой прямой:

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg = False)
```

кастомизация:

```python
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg = False)
ax.set_xlabels('Math score') #изменение названия оси
ax.set_ylabels('Reading score') #изменение названия оси
```

# Варианты команд для графиков

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('income.csv')
#варианты постройки графика
df.income.plot() #matplotlib
sns.lineplot(data=df)  #matplotlib
sns.lineplot(x=df.index, y=df.income) #matplotlib
df.plot() #matplotlib
df['income'].plot() #matplotlib
plt.plot(df.index, df.income)  #matplotlib
df.plot(kind='line') #matplotlib
```

```python
df = pd.read_csv("dataset_209770_6.txt", sep=" ")
sns.lmplot(x='x', y='y', data=df) #график из библиотеки seaborn
```

```python
df = pd.read_csv("genome_matrix.csv", index_col = 0)
f, ax = plt.subplots(figsize=(9, 6))
g = sns.heatmap(data = df, annot=True, linewidths=.1 , ax=ax, cmap='viridis')
g.xaxis.set_ticks_position('top')
g.xaxis.set_tick_params(rotation=90)
```

```python
df = pd.read_csv("genome_matrix.csv", index_col = 0)
g = sns.heatmap(data = df, cmap='viridis')
g.xaxis.set_ticks_position('top')
g.xaxis.set_tick_params(rotation=90)
```

```python
genome = pd.read_csv('genome_matrix.csv', index_col=0)
g = sns.heatmap(genome, cmap='viridis')
g.xaxis.set_ticks_position('top')
g.xaxis.set_tick_params(rotation=90)
```
