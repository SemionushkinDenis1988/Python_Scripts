```python
from time import time
before = time()
df.apply('mean')
after = time()
print(after - before)
```

```python
from time import time
before = time()
df.apply(np.mean)
after = time()
print(after - before)
```

```python
from time import time
before = time()
df.describe().loc['mean']
after = time()
print(after - before)
```
